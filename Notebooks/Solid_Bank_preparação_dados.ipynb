## **Feature engineering**

---



Lidando com dados temporais como idade e tempo de trabalho.

os dados de data de nascimento e anos empregados não estão em formatos adequados de datas. no nosso conjuto de dados os dados estão como contagem regressiva a partir do dia atual (0), -1 significa ontem. por isso é necessario uma transformação para conseguirmos extrair os anos desejados


```python
def data_nascimento(total_dias):
    hoje = datetime.date.today()
    aniversario = (hoje + timedelta(days=total_dias)).strftime('%Y-%m-%d')
    return aniversario                  

def dias_empregado(total_dias):
    hoje = datetime.date.today()
    adimissao = (hoje + datetime.timedelta(days=total_dias)).strftime('%Y-%m-%d')
    return adimissao

def idade(data_nascimento):
    hoje=datetime.date.today()
    data_de_nascimento=datetime.datetime.strptime(data_nascimento,'%Y-%m-%d')
    idade=hoje.year-data_de_nascimento.year
    return idade

def anos_emprego(data_empregado):
    hoje=datetime.date.today()
    data_emprego=datetime.datetime.strptime(data_empregado,'%Y-%m-%d')
    anos=hoje.year-data_emprego.year
    return anos
```

```python
df_cadastros_port_outl['DATA_NASCIMENTO']=df_cadastros_port_outl['DATA_NASCIMENTO'].apply(data_nascimento)
```

```python
df_cadastros_port_outl['DATA_EMPREGADO']=df_cadastros_port_outl['DATA_EMPREGADO'].apply(dias_empregado)
```

```python
df_cadastros_port_outl['IDADE']=df_cadastros_port_outl['DATA_NASCIMENTO'].apply(idade)
```

```python
df_cadastros_port_outl['ANOS_EMPREGO']=df_cadastros_port_outl['DATA_EMPREGADO'].apply(anos_emprego)
```

A partir disso ja podemos excluir as colunas de `DATA_NASCIMENTO` e `DATA_EMPREGADO`

# **Dados de apovação**

---

Dados referente ao **status de credito** de um cliente especifico em vários meses que aquele cliente solicitou um emprestimo ao banco.


Nesse banco de dados temos varios clientes com mais de uma obsevação pois o mesmo cliente pode solicitar emprestimos em varios meses diferentes, com isso temos o status de pagamento de cada mês específico para cada cliente.

`ID:`	Número do cliente.

`MONTHS_BALANCE:`		O mês dos dados extraídos é o ponto de partida, para trás, 0 é o mês atual, -1 é o mês anterior e assim por diante.

`STATUS:`	0: 1 a 29 dias de atraso 1: 30 a 59 dias de atraso 2: 60 a 89 dias de atraso 3: 90 a 119 dias de atraso 4: 120 a 149 dias de atraso 5: Vencido ou incobrável, baixas por mais de 150 dias C: pago naquele mês X: Sem empréstimo no mês.

```python
df_status_pagamento.head()
```

A tarefa aqui é determinar quias clientes são BONS ou MAUS pagadores, por isso o status de atraso de pagamento referindo aos dias de atraso de cada cliente nao resolve nosso problemas. Precisamos criar entao janelas de tempo para decidir se um cliente é bom ou mau pagador. Bons pagadores serão aqueles clientes com pagamentos até 60 dias de atraso se enquadrando no status 2 do nosso data set. A partir de 60 dias de atraso consideramos como mau pagador.

consideramos até 60 dias de atraso pois bancos tambem ganham dinheiro com juros de atraso de pagamentos.

```python
df_status_pagamento['STATUS'].unique()
```

```python
list1 = ['X', '0', 'C', '1'] # lista de bons pagadores (pagadores com menos de 60 dias)
df_status_pagamento['BOM_PAGADOR'] = df_status_pagamento.apply(lambda x: 1 if x.STATUS in list1 else 0, axis=1) #substituindo por 1 e 0 os bons e mau pagadores
```

```python
df_status_pagamento.drop(['STATUS'],axis=1,inplace=True)
df_status_pagamento.head()
```

Como existem diversos meses vamos determinar que se um cliente atrase mais de 60 dias em determinado mês esse cliente nao é um BOM pagador

```python

df = df_status_pagamento.groupby('ID')['BOM_PAGADOR'].mean().reset_index() # media de pagamento por cliente, caso a media seja "1" significa que ele sempre foi um bom pagador
df
```

```python
df['BOM'] = df.apply(lambda x: 1 if x['BOM_PAGADOR']==1 else 0, axis=1) # atribuindo valor "1" caso aquele cliente sempe foi bom pagador e "0" caso ja tenha atrasado alguma vez
```

```python
df[df['BOM_PAGADOR']!=1].value_counts().sum() 
df.drop('BOM_PAGADOR',axis=1,inplace=True)
```

Como podemos observar existem apenas `667` Maus pagadores em nosso conjunto de dados, o que nos diz que estamos lidando com dados extremamente desbalanceados. O balanceamento desse dataset será feito em outro notebook
